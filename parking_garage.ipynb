```python
class Garage():
    ticket_price = 5
    
    def __init__(self, avail_tickets = list(range(1,101)), avail_parking = list(range(1,101)), current_ticket = "", ticket_status = {"paid": False}):
        self.avail_tickets = avail_tickets
        self.avail_parking = avail_parking
        self.ticket_status = ticket_status
        self.current_ticket = current_ticket

    def takeTicket(self):
        if len(self.avail_parking) == 0:
            print("Sorry, lot is full. Please wait for car to exit.")
        else:
            self.current_ticket = self.avail_tickets.pop()
            self.avail_parking.pop()
            print(f"Here is your ticket. Ticket number " + str(self.current_ticket) )
            
    def payForParking(self):
        print(f"Your total for parking is ${self.ticket_price}.")
        payment = input("Please enter payment amount. ")
        if payment.isnumeric():
            if int(payment) == 5:
                print("Your ticket has been paid. You have 15 minutes to exit.")
                self.ticket_status["paid"] = True
            elif int(payment) < 5:
                print("You must pay total amount due. No payment transaction has processed.")
                self.payForParking()
            elif int(payment) > 5:
                change = int(payment) - 5
                print(f"Here is your change ${change}.")
                print("Your ticket has been paid. You have 15 minutes to exit.")
                self.ticket_status["paid"] = True
        else:
            print("That is not a form of payment we accept.")
            
    def leaveGarage(self):
        if self.ticket_status['paid'] == True:
            print("Thank you, have a nice day!")
            self.avail_parking.append(len(self.avail_parking))
            self.avail_tickets.append(self.current_ticket)
            self.ticket_status['paid'] = False
            self.current_ticket = ""
        else:
            print("You have not paid for parking.")
            self.payForParking()
            
my_garage = Garage()

def run_garage():
    print(f"Welcome to the parking spot. There are {len(my_garage.avail_parking)} spots availabe.")
    while True:
        action = input("Select an action ticket/pay or leave. ")
        if action == "ticket":
            my_garage.takeTicket()
        elif action == "pay":
            my_garage.payForParking()
        elif action == "leave":
            my_garage.leaveGarage()
            
run_garage()
```
